```python
image_list=[]#image listing array
y_train=[]#training pattern array
for x in glob.glob('/Users/yasha/Desktop/Spoof_data/resizedimages/*.png'):
    print(x)
    name=str(x).split('\\')[1]
    if(name[0]=='l'):
        y_train.append(1) #adding target varaible value to live(1) and spoof(0) images
        print(x,1)
    else:
        y_train.append(0)
        print(x,0)
    im=cv2.imread(x)
    print(im.shape)
    image_list.append(im) # shaping images into arrays
```

```python
imgs=[] #3d array violates LBP method 
i=0
for ims in image_list:
    ims=ims.reshape(3,-1)
    imgs.append(ims)
    i+=1#Shaping to 2d array
```

```python
imgs[1].shape 
```

```python
print(i)
```

```python
#extracting lbp features from each image
lbp=[]
for i in imgs:
    lbpt = feature.local_binary_pattern(i,24,8, method="uniform")
    (hist, _)=np.histogram(lbpt.ravel(),bins=np.arange(0, 27),range=(0,26))
    hist=hist.reshape(26,1)
    lbp.append(hist)
```

```python
lbp[2].shape
```

```python
#flattening the lbp features to feed them to the classifier
temp=1
for k in lbp:
    arr=np.array(k)
    arr_1d=arr.flatten()
    arr_2d=arr_1d.reshape(1,26)
    if(temp==1):
        X_img=arr_2d
        temp=0
    else:
        X_img=np.append(X_img,arr_2d,axis=0)
```
